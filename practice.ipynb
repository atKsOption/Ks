```python
class MuffledCalculator:
    muffled = False
    def calc(self,expr):
        try:
            return eval(expr)
        except ZeroDivisionError:
            if self.muffled:
                print 'Division by zero is illegal'
            else:
                raise
                
                
                
cal = MuffledCalculator()
cal.muffled = True
cal.calc('10/0')
```
